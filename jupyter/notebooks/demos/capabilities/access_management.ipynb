# Access management

## Access to management function with identity token

```python
from libs.helpers.utils import get_featuremesh_config

fm_config = get_featuremesh_config()

__YOUR_IDENTITY_TOKEN__ = fm_config['identity_token']
```

```python
import featuremesh
from featuremesh import AccessClient

featuremesh.set_default('access.host', fm_config['access.host'])

client_access = AccessClient(identity_token=__YOUR_IDENTITY_TOKEN__)
client_access.list_projects()
```

## Access to a project with an identity token

```python
__YOUR_IDENTITY_TOKEN__
```

```python
from featuremesh import generate_access_token

__YOUR_ACCESS_TOKEN__ = generate_access_token(
    identity_token=__YOUR_IDENTITY_TOKEN__, 
    project='default'
)
```

```python
from featuremesh import OfflineClient, Backend
import duckdb

featuremesh.set_default('registry.host', fm_config['registry.host'])
client_duckdb = OfflineClient(
    access_token=__YOUR_ACCESS_TOKEN__,
    backend=Backend.DUCKDB,
    sql_executor=lambda sql: duckdb.sql(sql).df()
)
client_duckdb.query('SELECT 1+2').dataframe
```

## Decode any token

```python
from featuremesh import decode_token

print("Identity token:\n", decode_token(__YOUR_IDENTITY_TOKEN__))
print("\nAccess token:\n", decode_token(__YOUR_ACCESS_TOKEN__))
```
